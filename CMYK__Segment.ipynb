```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt

def CMYK_color_image(image):
    img = image.astype(np.float64) / 255
    K = 1 - np.max(img, axis=2)
    C = (1 - img[..., 2] - K) / (1 - K)
    M = (1 - img[..., 1] - K) / (1 - K)
    Y = (1 - img[..., 0] - K) / (1 - K)

    CMYK_image = (np.dstack((C, M, Y, K)) * 255).astype(np.uint8)
    return CMYK_image

def label_and_save_images(directory_path, output_directory, label):
    image_paths = [os.path.join(directory_path, filename) for filename in os.listdir(directory_path)]

    for img_path in image_paths:
        #print(f"Processing image: {img_path}")
        img_OpenCV = cv2.imread(img_path, cv2.IMREAD_COLOR)
        CMYK_image = CMYK_color_image(img_OpenCV)

        # Get the name of the last directory in the path
        folder_name = os.path.basename(os.path.dirname(img_path))

        # Create the destination directory if it doesn't exist
        os.makedirs(os.path.join(output_directory, folder_name), exist_ok=True)

        # Save the segmented image with label to the corresponding destination folder
        output_path = os.path.join(output_directory, folder_name, f"{label}_{os.path.basename(img_path)}")
        cv2.imwrite(output_path, cv2.cvtColor(CMYK_image, cv2.COLOR_RGB2BGR))

# Set paths
train_dir = '/content/drive/MyDrive/XLA_S/Training'
output_CMYK_dir = '/content/drive/MyDrive/XLA_S/C_M_Y_K/Train'

# Process, save, and label caries images in the training directory
caries_train_dir = os.path.join(train_dir, 'caries')
label_and_save_images(caries_train_dir, output_CMYK_dir, label='Caries')

# Process, save, and label no caries images in the training directory
no_caries_train_dir = os.path.join(train_dir, 'No Caries')
label_and_save_images(no_caries_train_dir, output_CMYK_dir, label='No_Caries')
```

```python
def segment_color_image(image_path, threshold=100):
    # Read the color image using OpenCV
    img_OpenCV = cv2.imread(image_path, cv2.IMREAD_COLOR)
    # Create an image to store the segmentation result
    SegmentationIMG = Image.new("RGB", (img_OpenCV.shape[1], img_OpenCV.shape[0]))
    # Get image dimensions
    Width, Height = img_OpenCV.shape[1], img_OpenCV.shape[0]
    # Initialize cumulative values for average color
    a_TB = np.zeros(3, dtype=np.float)
    # Calculate average pixel value for the entire image
    for x in range(Width):
        for y in range(Height):
            pixel_value = img_OpenCV[y, x]
            a_TB += pixel_value
    # Calculate the size of the vector a: Size = Width * Height
    Size = Width * Height
    a_TB /= Size
    # Loop through all pixels in the image
    for a in range(Width):
        for b in range(Height):
            pixel_value = img_OpenCV[b, a]
            # Calculate Euclidean Distance D(z, a)
            #Tnh k/c D theo Euclidean Distance (6.7-1 /445)
            #D(z,a) = SQRT[(zR - aR)^2 + (zG - aG)^2 + (zB - aB)^2]
            D = np.linalg.norm(pixel_value - a_TB)
            # Compare with the threshold to determine if z(x, y) is background or object
            if D <= threshold:
                SegmentationIMG.putpixel((a, b), (255, 255, 255))  # White pixel
            else:
                SegmentationIMG.putpixel((a, b), tuple(pixel_value))
    # Convert the segmentation result from PIL to OpenCV for display
    Segmentation = np.array(SegmentationIMG)
    # Display the images using Colab's cv2_imshow
    cv2_imshow(img_OpenCV)
    cv2_imshow(Segmentation)

# Example usage
image_path = '/content/drive/MyDrive/XLA_S/C_M_Y_K/Train/caries/Caries_100.jpg'
segment_color_image(image_path)
```

```python
import os
import cv2
import numpy as np
from PIL import Image

# Hm segment_color_image

def segment_color_image(image, threshold=100):
    # Create an image to store the segmentation result
    SegmentationIMG = Image.new("RGB", (image.shape[1], image.shape[0]))
    # Initialize cumulative values for average color
    a_TB = np.zeros(3, dtype=np.float)
    # Calculate average pixel value for the entire image
    for x in range(image.shape[1]):
        for y in range(image.shape[0]):
            pixel_value = image[y, x]
            a_TB += pixel_value
    # Calculate the size of the vector a: Size = Width * Height
    Size = image.shape[1] * image.shape[0]
    a_TB /= Size
    # Loop through all pixels in the image
    for a in range(image.shape[1]):
        for b in range(image.shape[0]):
            pixel_value = image[b, a]
            # Calculate Euclidean Distance D(z, a)
            D = np.linalg.norm(pixel_value - a_TB)
            # Compare with the threshold to determine if z(x, y) is background or object
            if D <= threshold:
                SegmentationIMG.putpixel((a, b), (255, 255, 255))  # White pixel
            else:
                SegmentationIMG.putpixel((a, b), tuple(pixel_value))
    # Convert the segmentation result from PIL to OpenCV
    segmented_image = np.array(SegmentationIMG)
    return segmented_image

def save_segmented_images(directory_path, output_directory):
    image_paths = [os.path.join(directory_path, filename) for filename in os.listdir(directory_path)]
    for img_path in image_paths:
        img_OpenCV = cv2.imread(img_path, cv2.IMREAD_COLOR)
        # Convert BGR to RGB
        img_RGB = cv2.cvtColor(img_OpenCV, cv2.COLOR_BGR2RGB)

        segmented_image = segment_color_image(img_RGB)

        # Ly tn ca th mc cui cng trong ng dn
        folder_name = os.path.basename(os.path.dirname(img_path))

        # To th mc ch nu n khng tn ti
        output_folder_path = os.path.join(output_directory, folder_name)
        os.makedirs(output_folder_path, exist_ok=True)

        # Lu hnh nh sau khi phn on vo th mc ch tng ng
        output_path = os.path.join(output_folder_path, os.path.basename(img_path))
        cv2.imwrite(output_path, cv2.cvtColor(segmented_image, cv2.COLOR_RGB2BGR))

        # Phn loi hnh nh thnh "Caries" v "NoCaries" da trn tn th mc
        if "Caries" in folder_name:
            caries_output_path = os.path.join(output_directory, "Caries", os.path.basename(img_path))
            cv2.imwrite(caries_output_path, cv2.cvtColor(segmented_image, cv2.COLOR_RGB2BGR))
        elif "NoCaries" in folder_name:
            no_caries_output_path = os.path.join(output_directory, "NoCaries", os.path.basename(img_path))
            cv2.imwrite(no_caries_output_path, cv2.cvtColor(segmented_image, cv2.COLOR_RGB2BGR))

# ng dn th mc cha hnh nh gc
input_directory1 = '/content/drive/MyDrive/XLA_S/C_M_Y_K/Train/caries'
input_directory2 = '/content/drive/MyDrive/XLA_S/C_M_Y_K/Train/No Caries'

# Th mc u ra cha hnh nh  c phn loi
output_directory = "/content/drive/MyDrive/XLA_S/C_M_Y_K/Segment/Train"

# Gi hm  phn loi v lu hnh nh
save_segmented_images(input_directory1, output_directory)
save_segmented_images(input_directory2, output_directory)
```

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from keras.optimizers import Adam
from keras.regularizers import l2
from keras.preprocessing.image import ImageDataGenerator

train_dir = '/content/drive/MyDrive/XLA_S/C_M_Y_K/Segment/Train'
batch_size = 256
input_shape = (224, 224, 3)

train_datagen = ImageDataGenerator(
    rescale=1.0/255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest')

test_datagen = ImageDataGenerator(rescale=1.0/255)

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(224, 224),
    batch_size=batch_size,
    class_mode='binary')  # Assuming you have binary classification, change it if you have more classes

# Assuming you have a 'test_dir' directory for testing
test_dir = '/content/drive/MyDrive/XLA_S/C_M_Y_K/Segment/Test'
test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(224, 224),
    batch_size=batch_size,
    class_mode='binary')  # Assuming you have binary classification, change it if you have more classes

model = Sequential()
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=input_shape))
model.add(MaxPooling2D((2, 2)))
model.add(Flatten())
model.add(Dense(128, activation='relu', kernel_regularizer=l2(0.01)))
model.add(Dropout(0.5))
model.add(Dense(1, activation='sigmoid'))  # Assuming you have binary classification, change it if you have more classes

# Compile model
model.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
history = model.fit(
    train_generator,
    steps_per_epoch=len(train_generator),
    epochs=20,
    validation_data=test_generator,
    validation_steps=len(test_generator)
)
# Model evaluation
test_loss, test_acc = model.evaluate(test_generator, steps=len(test_generator))
print(f"Test accuracy: {test_acc:.2f}")
```
